```python
import pandas as pd
import numpy as np

# load the data set
# note this data uses tab (\t) separator
df = pd.read_csv("/content/marketing_campaign.csv", sep="\t")
```

```python
df.head()
```

```python
## to understand the dataset structure.
## to identify the nulls ,data types, numeric vs categorical columns.

df.shape
df.info
df.describe(include='all')
```

```python
# check the missing valuse
df.isnull().sum()
```

```python
#let's fill the income  missing values with median
df['Income'] = df['Income'].fillna(df['Income'].median())

```

```python
print(df['Income'].value_counts())
```

```python
# check duplicates and remove duplicates is have any
df.duplicated().sum()
df.drop_duplicates(inplace = True)

```

```python
# clean and standardize categorical columns
# stripe and lowercase marital status
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()

# simplify similar categories (e.g, absurd, yolo -> single)
df['Marital_Status'] = df['Marital_Status'].replace({
   'divorced': 'single', 'widow': 'single',
    'alone': 'single', 'absurd': 'single',
    'yolo': 'single'
})
```

```python
print(df['Marital_Status'].value_counts())
```

```python
##convert enrollment date to datetime

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
print(df['Dt_Customer'].value_counts())
```

```python
# Rename all columns to meaningful and consistent names
df.rename(columns={
    'ID': 'customer_id',
    'Year_Birth': 'year_of_birth',
    'Education': 'education_level',
    'Marital_Status': 'marital_status',
    'Income': 'income',
    'Kidhome': 'num_kids_home',
    'Teenhome': 'num_teens_home',
    'Dt_Customer': 'enrollment_date',
    'Recency': 'days_since_last_purchase',

    'MntWines': 'spend_wine',
    'MntFruits': 'spend_fruits',
    'MntMeatProducts': 'spend_meat',
    'MntFishProducts': 'spend_fish',
    'MntSweetProducts': 'spend_sweets',
    'MntGoldProds': 'spend_gold',

    'NumDealsPurchases': 'purchases_with_discount',
    'NumWebPurchases': 'purchases_web',
    'NumCatalogPurchases': 'purchases_catalog',
    'NumStorePurchases': 'purchases_store',
    'NumWebVisitsMonth': 'website_visits_month',

    'AcceptedCmp1': 'campaign_1_accepted',
    'AcceptedCmp2': 'campaign_2_accepted',
    'AcceptedCmp3': 'campaign_3_accepted',
    'AcceptedCmp4': 'campaign_4_accepted',
    'AcceptedCmp5': 'campaign_5_accepted',

    'Response': 'last_campaign_response'
}, inplace=True)

# View the new column names
print(df.columns.tolist())
```

```python
#validate and fix data types
# convert year_of_birth to int

df['year_of_birth'] = df['year_of_birth'].astype(int)

# confirn the data types now
df.dtypes
```

```python
# Save cleaned data
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
df = pd.read_csv("cleaned_marketing_campaign.csv")

# Show basic info again
df.info()
```

```python
# age distribution
#create a column foe age
df['age'] = 2025 - df['year_of_birth']

# plot age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True, color='teal')
plt.title('AGE DISTRIBUTION')
plt.xlabel('AGE')
plt.ylabel('NUMBER OF CUSTOMER')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='marital_status',hue='marital_status', data=df, palette='Set2')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['income'], bins=30, kde=True, color='purple')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.show()
```

```python
# Total amount spent by each customer
df['total_spent'] = df[[
    'spend_wine', 'spend_fruits', 'spend_meat', 'spend_sweets',
    'spend_gold', 'spend_fish'
]].sum(axis=1)

# Spending vs Income
plt.figure(figsize=(8,5))
sns.scatterplot(x='income', y='total_spent', data=df, alpha=0.6)
plt.title('Income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```

```python
df['enrollment_year'] = pd.to_datetime(df['enrollment_date']).dt.year

plt.figure(figsize=(6,4))
sns.countplot(x='enrollment_year', hue='enrollment_year', data=df, palette='coolwarm', legend=False)
plt.title('Year of Customer Enrollment')
plt.xlabel('Year')
plt.ylabel('Customers Joined')
plt.show()
```
